# Format Strings -> fmt

we can also use the shortcut notation parameter to specify the marker.
The parameter is also called "fmt". and is written with this syntax -

marker | line | color

for line - \
"-" -> solid line \
":" -> dotted line  \
"--" -> dashed line  \
"-." -> dashed dotted line \

for color - \
r, g, c, g, m, y, k, w

```python
plt.plot(ypoint, "*--b")
plt.show()
```

to increase marker size -> use "ms" or "markersize" = any int valuue.

```python
plt.plot(ypoint, "*--b", ms = 10)
plt.show()
```

to give marker color -> mec or markeredgecolor = "k", mfc or markerfacecolor = "w".

```python
plt.plot(ypoint, "*-b", mfc = "g", mec = "g", ms = 20)
plt.show()
```

there are over 140 supported colors. we can give it by directly passing color name.

```python
plt.plot(ypoint, "*:b", mfc = "LightCyan", mec = "c", ms = 20)
plt.show()
```

for more colors -> https://www.w3schools.com/colors/colors_names.asp

# Subplot

is used to draw multiple plots in same figure.

plt.subplot() -> takes 3 arguments that describes the layout of the figure. \
plt.subplot(a, b, c) -> a rows, b columns and cth plot.

```python
x = np.array([1,4,7,9])
y = np.array([4,6,5,2])

# plot 1
plt.subplot(1, 2, 1)
plt.title("Graph 1")
plt.plot(x, y)


x = np.array([5,4,3,2,1])
y = np.array([1,2,3,4,5])

# plot 2
plt.subplot(1, 2, 2)
plt.title("Graph 2")
plt.plot(x, y)


plt.show()
```

giving super title with -> suptitle() function

```python
x = np.array([1,2,3,4])
y = np.array([1,2,3,4])

# plot 1
plt.subplot(1, 2, 1)
plt.title("Product 1")
plt.plot(x, y)


x = np.array([5,4,3,2,1])
y = np.array([1,2,3,4,5])

# plot 2
plt.subplot(1, 2, 2)
plt.title("Product 2")
plt.plot(x, y)

plt.suptitle("My Shop")
plt.show()
```

ColorMap -> Matplot have number of colormaps available. A colormap is a list of colors where each color ranges from 0 to 100.

```python
# using colormap

x = np.random.randint(10, size = 50)
y = np.random.randint(10, size = 50)

colors = np.random.randint(100, size = 50)

plt.scatter(x, y, c = colors, cmap = "Blues")

# for showing colorb ar
plt.colorbar()

plt.show()
```

```python
# using colormap

x = np.random.randint(10, size = 50)
y = np.random.randint(10, size = 50)

colors = np.random.randint(100, size = 50)
sizes = np.random.randint(100, size = 50)

plt.scatter(x, y, c = colors, cmap = "Blues", s = sizes)

# for showing colorb ar
plt.colorbar()

plt.show()
```

# Pie Chart

the plotting of pie chart by default starts from x axis and moves counterclockwise.

```python
p = np.array([25, 40, 15, 20, 27])

plt.pie(p)
plt.show()
```

```python
p = np.array([25, 40, 15, 20, 27])
my_labels = ["Apple", "Mango", "Banana", "Litchi", "Chiku"]

plt.pie(p, labels = my_labels)
plt.show()
```

chaging the start angle.

```python
p = np.array([25, 40, 15, 20, 27])
my_labels = ["Apple", "Mango", "Banana", "Litchi", "Chiku"]

plt.pie(p, labels = my_labels, startangle = 45)
plt.show()
```

to increase pie chart size.

```python
p = np.array([25, 40, 15, 20, 27])
my_labels = ["Apple", "Mango", "Banana", "Litchi", "Chiku"]

plt.pie(p, labels = my_labels, startangle = 45, radius = 1.5)
plt.show()
```

Explode -> used to stand out wedges.

```python
p = np.array([25, 40, 15, 20, 27])
my_labels = ["Apple", "Mango", "Banana", "Litchi", "Chiku"]

plt.pie(p, labels = my_labels, startangle = 45, explode = [0, 0.1, 0, 0, 0])
plt.show()
```

```python
p = np.array([25, 40, 15, 20, 27])
my_labels = ["Apple", "Mango", "Banana", "Litchi", "Chiku"]

# adding shadow to out wedges.
plt.pie(p, labels = my_labels, startangle = 45, explode = [0, 0.1, 0, 0, 0], shadow = True)
plt.show()
```

```python
p = np.array([25, 40, 15, 20, 27])
my_labels = ["Apple", "Mango", "Banana", "Litchi", "Chiku"]

# adding shadow to out wedges.
plt.pie(p, labels = my_labels, explode = [0, 0.1, 0, 0, 0], shadow = True)
plt.legend()
plt.show()
```

adding title to the legend.

```python
p = np.array([25, 40, 15, 20, 27])
my_labels = ["Apple", "Mango", "Banana", "Litchi", "Chiku"]

# adding shadow to out wedges.
plt.pie(p, labels = my_labels, explode = [0, 0.1, 0, 0, 0], shadow = True)
plt.legend(title = "Fruits detail:")
plt.show()
```

```python
p = np.array([25, 40, 15, 20, 27])
my_labels = ["Apple", "Mango", "Banana", "Litchi", "Chiku"]

# adding shadow to out wedges.
plt.pie(p, labels = my_labels, explode = [0, 0.1, 0, 0, 0], shadow = True)
plt.legend(title = "Fruits detail:")
plt.axis("equal")
plt.show()
```

adding percentage info into the chart.

```python
p = np.array([25, 40, 15, 20, 27])
my_labels = ["Apple", "Mango", "Banana", "Litchi", "Chiku"]

# adding shadow to out wedges.
plt.pie(p, labels = my_labels, explode = [0, 0.1, 0, 0, 0], shadow = True, autopct = "%0.0f%%")
plt.legend(title = "Fruits detail:")
plt.axis("equal")
plt.show()
```

```python
p = np.array([25, 40, 15, 20, 27])
my_labels = ["Apple", "Mango", "Banana", "Litchi", "Chiku"]

# adding shadow to out wedges.
plt.pie(p, labels = my_labels, explode = [0, 0.1, 0, 0, 0], shadow = True, autopct = "%0.2f%%")
plt.legend(title = "Fruits detail:")
plt.axis("equal")
plt.show()
```
